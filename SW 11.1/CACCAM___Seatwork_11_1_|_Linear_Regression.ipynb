- PART 1

```python
# Given values
b = 1
w1 = 3
x1 = 0

# Linear regression prediction: y = w1*x1 + b
y = w1 * x1 + b

print("Prediction:", y)
```

Analysis:
- The prediction is 1 because the input 𝑥 1 = 0, so the term 𝑤 1 ⋅ 𝑥 1  becomes 0. That means the output depends only on the bias 𝑏 . So, the model outputs the bias value, which is 1.

- PART 2

```python
# Given weights and bias
b = 1
w1 = 3
w2 = -2
w3 = 4

# Example 1
x1_1, x2_1, x3_1 = 3, -1, 2
y1 = w1*x1_1 + w2*x2_1 + w3*x3_1 + b

# Example 2
x1_2, x2_2, x3_2 = 10, 0, -10
y2 = w1*x1_2 + w2*x2_2 + w3*x3_2 + b

print("Prediction for Example 1:", y1)
print("Prediction for Example 2:", y2)

```

Analysis:
- 20 for Example 1 and -9 for Example 2. Example 1 is positive because all features contribute positively overall, while Example 2 becomes negative due to the strong negative effect of x 3 ​
