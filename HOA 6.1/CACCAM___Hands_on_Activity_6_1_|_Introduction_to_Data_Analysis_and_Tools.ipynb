### Exercise 1

```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
len(salaries)
```

```python
# Mean

sum(salaries)/len(salaries)
```

```python
# Median

sorted_salaries = sorted(salaries)

n = len(sorted_salaries)
mid = n // 2

if n % 2 == 1:
    median = sorted_salaries[mid]
else:
    median = (sorted_salaries[mid - 1] + sorted_salaries[mid]) / 2

print(median)
```

```python
# Mode

map_values = {}
for salary in salaries:
  if salary not in map_values:
    map_values[salary] = 1
    continue
  map_values[salary] += map_values[salary] + 1

print(map_values)
```

```python
# Standard Variance

average = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1

x_xmean_squared = []
for salary in salaries:
  temp = (salary - average)**2
  x_xmean_squared.append(temp)

summation = sum(x_xmean_squared)
variance = summation/len_minus_one

print(variance)
```

```python
# Standard Deviation

squared_diffs = [(x - mean) ** 2 for x in salaries]
variance = sum(squared_diffs) / (len(salaries) - 1)
std_dev = variance ** 0.5

print(std_dev)
```

```python
import statistics
```

```python
# Range

range = max(salaries) - min(salaries)
print(range)
```

```python
# Coefficient of variation

data = salaries
mean = statistics.mean(salaries)
std_dev = statistics.stdev(salaries)

print((std_dev - mean)*100)
```

```python
# Calculate the Interquartile Range

mid_index = n // 2
if n % 2 == 0:
    lower_half = data[:mid_index]
    upper_half = data[mid_index:]
else:
    lower_half = data[:mid_index]
    upper_half = data[mid_index+1:]

Q1 = statistics.median(lower_half)
Q3 = statistics.median(upper_half)

print(Q1 - Q3)
```

```python
# Quartile Coefficient of Dispersion

mid_index = n // 2
if n % 2 == 0:
    lower_half = data[:mid_index]
    upper_half = data[mid_index:]
else:
    lower_half = data[:mid_index]
    upper_half = data[mid_index+1:]

Q1 = statistics.median(lower_half)
Q3 = statistics.median(upper_half)

print((Q3 - Q1) / (Q3 + Q1))
```
