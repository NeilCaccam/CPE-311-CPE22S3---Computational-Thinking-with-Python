## Example of gathering image data using webcam

```python
import cv2

webcam = cv2.VideoCapture(0)

try:
    while True:
        check, frame = webcam.read()
        print(check) #prints true as long as the webcam is running
        print(frame) #prints matrix values of each frame
        cv2.imshow("Capturing", frame)
        key = cv2.waitKey(1)
        
        if key == ord('s'):
            cv2.imwrite(filename='saved_img.jpg', img=frame)
            webcam.release()
            img_new = cv2.imread('saved_img.jpg', cv2.IMREAD_GRAYSCALE)
            cv2.imshow("Captured Image", img_new)
            cv2.waitKey(1650)
            cv2.destroyAllWindows()
            print("Processing image...")
            img_ = cv2.imread('saved_img.jpg', cv2.IMREAD_ANYCOLOR)
            print("Converting RGB image to grayscale...")
            gray = cv2.cvtColor(img_, cv2.COLOR_BGR2GRAY)
            print("Converted RGB image to grayscale...")
            print("Resizing image to 28x28 scale...")
            img_ = cv2.resize(gray,(28,28))
            print("Resized...")
            cv2.imwrite(filename='saved_img-final.jpg', img=img_)
            print("Image saved!")
            break
            
        elif key == ord('q'):
            print("Turning off camera.")
            webcam.release()
            print("Camera off.")
            print("Program ended.")
            cv2.destroyAllWindows()
            break

except KeyboardInterrupt:
    print("Turning off camera.")
    webcam.release()
    print("Camera off.")
    print("Program ended.")
    cv2.destroyAllWindows()
```

```python
# import required libraries
import sounddevice as sd
from scipy.io.wavfile import write
import wavio as wv

# Sampling frequency
freq = 44100

# Recording duration
duration = 5

# Start recorder with the given values
# of duration and sample frequency
recording = sd.rec(int(duration * freq),
samplerate=freq, channels=2)

# Record audio for the given number of seconds
sd.wait()

# This will convert the NumPy array to an audio
# file with the given sampling frequency
write("recording0.wav", freq, recording)

# Convert the NumPy array to audio file
wv.write("recording1.wav", recording, freq, sampwidth=2)
```

```python
# Install required packages first in VS Code terminal:
# pip install selenium pillow requests webdriver-manager

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import requests
import os
import io
from PIL import Image
import urllib.parse
import base64

# Setup Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--window-size=1920,1080')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

# Add user agent to avoid detection
chrome_options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36')

# Setup driver with webdriver-manager
try:
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
    print(" Chrome driver setup successful!")
except Exception as e:
    print(f" Error setting up Chrome driver: {e}")
    exit()

def get_image_urls(search_term, max_images=5):
    """Get image URLs from Google Images"""
    # Encode search term for URL
    encoded_term = urllib.parse.quote(search_term)
    
    # URL for Google Images
    url = f"https://www.google.com/search?q={encoded_term}&tbm=isch&hl=en"
    
    print(f" Searching for: {search_term}")
    driver.get(url)
    time.sleep(3)
    
    image_urls = []
    attempts = 0
    processed_urls = set()
    
    # Scroll to load more images
    for _ in range(3):  # Scroll a few times to load more images
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    
    while len(image_urls) < max_images and attempts < 15:
        try:
            # Method 1: Get images from the src attribute of thumbnail images
            thumbnails = driver.find_elements(By.CSS_SELECTOR, "img.rg_i")
            
            for i, thumbnail in enumerate(thumbnails):
                if len(image_urls) >= max_images:
                    break
                
                try:
                    # Try to get the image source
                    # Method A: Check src attribute
                    src = thumbnail.get_attribute('src')
                    if src and src.startswith('http') and src not in processed_urls:
                        processed_urls.add(src)
                        image_urls.append(src)
                        print(f"  Found image {len(image_urls)}/{max_images} (from thumbnail)")
                        continue
                    
                    # Method B: Check data-src attribute
                    data_src = thumbnail.get_attribute('data-src')
                    if data_src and data_src.startswith('http') and data_src not in processed_urls:
                        processed_urls.add(data_src)
                        image_urls.append(data_src)
                        print(f"  Found image {len(image_urls)}/{max_images} (from data-src)")
                        continue
                    
                    # Method C: Click on thumbnail to get full image
                    if len(image_urls) < max_images:
                        driver.execute_script("arguments[0].scrollIntoView();", thumbnail)
                        time.sleep(0.5)
                        driver.execute_script("arguments[0].click();", thumbnail)
                        time.sleep(2)
                        
                        # Look for full-size image
                        full_images = driver.find_elements(By.CSS_SELECTOR, "img[jsname='HiaYvf'], img.n3VNCb, img.r48jcc")
                        for full_img in full_images:
                            full_src = full_img.get_attribute('src')
                            if full_src and full_src.startswith('http') and full_src not in processed_urls:
                                processed_urls.add(full_src)
                                image_urls.append(full_src)
                                print(f"   Found image {len(image_urls)}/{max_images} (after click)")
                                break
                                
                except Exception as e:
                    continue
            
            # Method 2: Try to get all images directly from the page
            all_images = driver.find_elements(By.TAG_NAME, "img")
            for img in all_images:
                if len(image_urls) >= max_images:
                    break
                    
                for attr in ['src', 'data-src', 'data-iurl']:
                    try:
                        url = img.get_attribute(attr)
                        if url and url.startswith('http') and url not in processed_urls:
                            # Filter out small icons and logos
                            if 'logo' not in url.lower() and 'icon' not in url.lower():
                                processed_urls.add(url)
                                image_urls.append(url)
                                print(f"   Found image {len(image_urls)}/{max_images} (direct)")
                                break
                    except:
                        continue
            
            # Method 3: Try to load more results
            try:
                # Click on "Show more results" button if it exists
                show_more = driver.find_element(By.CSS_SELECTOR, ".LZ4I, .mye4qd")
                driver.execute_script("arguments[0].click();", show_more)
                time.sleep(3)
                # Scroll after clicking
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)
            except:
                # If no "show more" button, just scroll
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)
            
        except Exception as e:
            print(f"   Error in extraction: {str(e)[:50]}")
        
        attempts += 1
    
    # Remove duplicates while preserving order
    seen = set()
    unique_urls = []
    for url in image_urls:
        if url not in seen:
            seen.add(url)
            unique_urls.append(url)
    
    print(f"  Total unique images found: {len(unique_urls)}")
    return unique_urls[:max_images]

def download_image(url, folder_path, file_name):
    """Download and save image from URL"""
    try:
        # Skip if it's a base64 encoded image
        if url.startswith('data:image'):
            print(f"   Skipping base64 image: {file_name}")
            return False
        
        # Download image
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
            'Accept': 'image/webp,image/apng,image/*,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.9',
            'Referer': 'https://www.google.com/'
        }
        
        response = requests.get(url, headers=headers, timeout=15, stream=True)
        response.raise_for_status()
        
        # Check content type
        content_type = response.headers.get('content-type', '')
        if 'image' not in content_type.lower():
            print(f"   Not an image: {content_type}")
            return False
        
        # Convert to image
        img = Image.open(io.BytesIO(response.content))
        
        # Convert to RGB if necessary
        if img.mode != 'RGB':
            img = img.convert('RGB')
        
        # Save image
        os.makedirs(folder_path, exist_ok=True)
        file_path = os.path.join(folder_path, file_name)
        img.save(file_path, 'JPEG', quality=85)
        
        # Check if file was saved and has size > 0
        if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
            print(f"   Saved: {file_name} ({os.path.getsize(file_path)/1024:.1f} KB)")
            return True
        else:
            print(f"   Saved file is empty: {file_name}")
            return False
        
    except Exception as e:
        print(f"   Failed to download {file_name}: {str(e)[:100]}")
        return False

def main():
    # Configuration
    search_terms = ['cat', 'dog']  # Add your search terms here
    images_per_term = 5
    
    # Create base download folder
    base_folder = os.path.join(os.path.expanduser("~"), "Downloads", "GoogleImages")
    print(f"Images will be saved to: {base_folder}")
    
    for term in search_terms:
        print(f"\n{'='*50}")
        print(f"Processing: {term}")
        print(f"{'='*50}")
        
        # Create folder for this search term
        term_folder = os.path.join(base_folder, term)
        os.makedirs(term_folder, exist_ok=True)
        
        # Get image URLs
        print(f" Fetching image URLs for '{term}'...")
        image_urls = get_image_urls(term, images_per_term)
        
        if not image_urls:
            print(f"No image URLs found for '{term}'")
            continue
            
        print(f" Found {len(image_urls)} images for '{term}'")
        
        # Download each image
        successful = 0
        for i, url in enumerate(image_urls):
            file_name = f"{term}_{i+1:02d}.jpg"
            if download_image(url, term_folder, file_name):
                successful += 1
            time.sleep(1)  # Be nice to servers
        
        print(f"Downloaded {successful}/{len(image_urls)} images for '{term}'")
    
    print(f"\n{'='*50}")
    print(f"All done! Images saved to: {base_folder}")
    print(f"{'='*50}")

# Run the script
if __name__ == "__main__":
    try:
        main()
        input("\nPress Enter to close the browser...")  # Wait for user input
    finally:
        # Always close the browser
        driver.quit()
        print("🔚 Browser closed")
```

```python
# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
metascores = []
votes = []

# Extract data from individual movie container
for container in movie_containers:
    
    # The name - clean it
    name = container.h3.text
    # Remove numbering (e.g., "1. Logan" -> "Logan")
    if '. ' in name:
        name = name.split('. ', 1)[1]
    names.append(name)
    print(f"Movie: {name}")
    
    # The year - handle both single years and ranges
    year_element = container.find('span', class_='sc-a55f6282-6 iMumIM dli-title-metadata-item')
    if year_element:
        year_text = year_element.text
        # Handle date ranges like "2017–2021" - take the first year
        if '–' in year_text:
            year = int(year_text.split('–')[0])
        elif '-' in year_text:
            year = int(year_text.split('-')[0])
        else:
            year = int(year_text)
        years.append(year)
    else:
        years.append(None)
    
    # The IMDB rating
    rating_element = container.find('span', class_='ipc-rating-star--rating')
    if rating_element:
        rating = float(rating_element.text)
        imdb_ratings.append(rating)
    else:
        imdb_ratings.append(None)
    
    # The Metascore
    m_score_element = container.find('span', class_='sc-9fe7b0ef-0 hDuMnh metacritic-score-box')
    if m_score_element:
        m_score = int(m_score_element.text)
        metascores.append(m_score)
    else:
        metascores.append(None)
    
    # The number of votes (with K/M handling)
    vote_element = container.find('span', class_='ipc-rating-star--voteCount')
    if vote_element:
        vote_text = vote_element.text
        # Clean the text
        vote_text = vote_text.replace('(', '').replace(')', '').replace('<!--', '').replace('-->', '').strip()
        
        # Handle K (thousands) and M (millions)
        if 'K' in vote_text:
            vote_text = vote_text.replace('K', '')
            vote_count = int(float(vote_text) * 1000)
        elif 'M' in vote_text:
            vote_text = vote_text.replace('M', '')
            vote_count = int(float(vote_text) * 1000000)
        else:
            vote_count = int(vote_text)
        
        votes.append(vote_count)
    else:
        votes.append(None)

# Create DataFrame
import pandas as pd
test_df = pd.DataFrame({
    'movie': names,
    'year': years,
    'imdb': imdb_ratings,
    'metascore': metascores,
    'votes': votes
})

# Display the DataFrame
print(test_df.info())
print("\n")
print(test_df)
```
