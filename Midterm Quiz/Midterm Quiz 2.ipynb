## Phase 1: Extract
In this phase, i have installed the kagglehub library so i can call the dataset nd then be retrieved from the source. To follow the requirements i used the Kaggle API for the dataset

```python
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

path = kagglehub.dataset_download("supplejade/rt-iot2022real-time-internet-of-things")
print("dataset path:", path)

df = pd.read_csv(f"{path}/RT_IOT2022.csv")

print("RT_IOT2022.csv", df.shape)
df.head(5)
```

## Phase 2: Transform
In this phase i am preparing the data for data exploration, which is going to be crucial in data exploritory and the accuracy for the outputs.

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)

label = LabelEncoder()
df['proto_encoded'] = label.fit_transform(df['proto'])
df['service_encoded'] = label.fit_transform(df['service'])

print("Transformation, New Shape:", df.shape)
```
