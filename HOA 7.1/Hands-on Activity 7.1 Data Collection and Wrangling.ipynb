# Exercise 1

##### We want to look at data for Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks, but were given each as a seperate CSV file. Combine them into single file and store the dataframe of the FAANG data for the rest of the exercises.

```python
import pandas as pd
import os

tickers = ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOG']
faang_list = []

for ticker in tickers:
    file_name = f"{ticker.lower()}.csv"
    
    if os.path.exists(file_name):
        df = pd.read_csv(file_name)
        
        df['ticker'] = ticker
        faang_list.append(df)
    else:
        print(f"Warning: {file_name} not found in the current folder.")

faang = pd.concat(faang_list, ignore_index=True)
faang.to_csv('faang.csv', index=False)
faang.head()
```

```python
import pandas as pd
import time
import re
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options as EdgeOptions
from selenium.webdriver.common.by import By

edge_driver_path = r"C:\Users\Neil Caccam\CPE-311-CPE22S3---Computational-Thinking-with-Python\HOA 7.1\msedgedriver.exe"

def run_st_lukes_edge_project():
    print("Step 1: Scraping St. Luke's data using Microsoft Edge...")
    
    options = EdgeOptions()
    options.add_argument("--remote-allow-origins=*")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36")
    options.add_argument("--start-maximized")

    service = EdgeService(executable_path=edge_driver_path)
    driver = webdriver.Edge(service=service, options=options)
    
    url = "https://www.stlukes.com.ph/contact-us"
    scraped_data = []

    try:
        driver.get(url)
        print("Waiting for page elements to load...")
        time.sleep(10) 

        location_blocks = driver.find_elements(By.CLASS_NAME, "contact-details")
        
        if not location_blocks:
            raise Exception("No data found on the page. Check if the class name 'contact-details' has changed.")

        for block in location_blocks:
            content = block.text.strip()
            if content:
                parts = content.split('\n')
                if len(parts) >= 3:
                    scraped_data.append([parts[0], parts[1], parts[2]])
        
        print(f"Successfully scraped {len(scraped_data)} locations!")

    except Exception as e:
        print(f"Connection/Scraping issue: {e}")
        print("Using fallback data...")
        scraped_data = [
            ["ST. LUKE'S MEDICAL CENTER - QUEZON CITY", "279 E Rodriguez Sr. Ave. Quezon City, 1112", "+63-2-8723-0101"],
            ["ST. LUKE'S MEDICAL CENTER- GLOBAL CITY", "Rizal Drive cor. 32nd St. and 5th Ave., Taguig, 1634", "+63-2-8789-7700"],
            ["ST. LUKE'S MEDICAL CENTER EXTENSION CLINIC", "1177 J. Bocobo Street, Ermita, Manila", "+63-2-8521-0020 / +63-2-8521-8647"]
        ]
    finally:
        driver.quit()

    df_raw = pd.DataFrame(scraped_data, columns=['Hospital_Name', 'Address', 'Phone'])
    df_raw.to_csv('hospitals.csv', index=False)
    
    df = pd.read_csv('hospitals.csv')
    df['Primary_Phone'] = df['Phone'].apply(lambda x: str(x).split('/')[0].strip())
    df['Phone_Digits'] = df['Primary_Phone'].apply(lambda x: re.sub(r'\D', '', x))
    df['Hospital_Name'] = df['Hospital_Name'].str.title()
    
    def get_city(address):
        parts = address.split(',')
        if len(parts) > 1:
            city_part = parts[-1].strip()
            city_cleaned = re.sub(r'\d{4}', '', city_part).strip()
            return city_cleaned if city_cleaned else "Manila"
        return "Unknown"

    df['City'] = df['Address'].apply(get_city)

    print("\n--- Final Data Output ---")
    print(df[['Hospital_Name', 'City', 'Phone_Digits']])
    df.to_csv('hospitals_processed.csv', index=False)

run_st_lukes_edge_project()
```
