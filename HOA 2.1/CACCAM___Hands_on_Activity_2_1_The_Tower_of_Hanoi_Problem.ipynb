#**Procedure**


**Observation**

In this part i have initialized a setup the stands on the tower of hanoi as (Source, Auxiliary, Destination) and also initialized values of the list to act as the disk in the game

i also added a feature that sorts the values as descending order to follow the game's rules (from biggest to smallest)

```python
source = [40, 30, 10, 20]
auxiliary = []
destination = []

source.sort(reverse = True)

print(source)
```

**Observation**

i defined a hanoi function this recursive structure ensures that disks are always moved in a way that a larger disk is never placed on top of a smaller disk, solving the Tower of Hanoi puzzle.

```python
#check the items and transfer to desitnation list

#else n-1 case

def hanoi(n, source, auxiliary, destination):
    if n == 1:
        # Move one disk directly
        destination.append(source.pop())
        print(f"Move disk to destination: {destination}")
    else:
        # Step 1: Move n-1 disks from source to auxiliary
        hanoi(n-1, source, destination, auxiliary)

        # Step 2: Move the largest disk from source to destination
        destination.append(source.pop())
        print(f"Move disk to destination: {destination}")

        # Step 3: Move n-1 disks from auxiliary to destination
        hanoi(n-1, auxiliary, source, destination)
```

```python
#run the hanoi function and the full source list

hanoi(len(source), source, auxiliary, destination)
```

```python
#print final output

print(source)
print(auxiliary)
print(destination)
```
