## 1. With the earthquakes.csv file, select all the earthquakes in Japan with a magType of mb and a magnitude of 4.9 or greater.

```python
import pandas as pd

earthquakes = pd.read_csv('earthquakes.csv')

japan_earthquakes = earthquakes[
    (earthquakes['parsed_place'] == 'Japan') & 
    (earthquakes['magType'] == 'mb') & 
    (earthquakes['mag'] >= 4.9)
]

print(japan_earthquakes)
```

## 2. Create bins for each full number of magnitude (for example, the first bin is 0-1, the second is 1-2, and so on) with a magType of ml and count how many are in each bin.

```python
import pandas as pd
import numpy as np

earthquakes = pd.read_csv('earthquakes.csv')

ml_earthquakes = earthquakes[earthquakes['magType'] == 'ml']

bins = np.arange(0, ml_earthquakes['mag'].max() + 1, 1)
labels = [f'{i}-{i+1}' for i in range(len(bins)-1)]

ml_earthquakes['mag_bin'] = pd.cut(ml_earthquakes['mag'], bins=bins, labels=labels, right=False)

bin_counts = ml_earthquakes['mag_bin'].value_counts().sort_index()
print(bin_counts)
```

## 3. Using the faang.csv file, group by the ticker and resample to monthly frequency. 

```python
import pandas as pd

faang = pd.read_csv('faang.csv')

faang['date'] = pd.to_datetime(faang['date'])

monthly_agg = faang.groupby('ticker').resample('ME', on='date').agg({
    'open': 'mean',
    'high': 'max',
    'low': 'min',
    'close': 'mean',
    'volume': 'sum'
}).round(2)

print(monthly_agg)
```

## 4. Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magType along the columns

```python
import pandas as pd

earthquakes = pd.read_csv('earthquakes.csv')

crosstab = pd.crosstab(
    index=earthquakes['tsunami'],
    columns=earthquakes['magType'],
    values=earthquakes['mag'],
    aggfunc='max'
).round(2)

print(crosstab)
```

## 5. Calculate the rolling 60-day aggregations of OHLC data by ticker for the FAANG data. Use the same aggregations as exercise no. 3.

```python
import pandas as pd

faang = pd.read_csv('faang.csv')
faang['date'] = pd.to_datetime(faang['date'])

faang = faang.set_index('date').sort_index()

rolling_60day = faang.groupby('ticker').rolling(window='60D').agg({
    'open': 'mean',
    'high': 'max',
    'low': 'min',
    'close': 'mean',
    'volume': 'sum'
}).round(2)

print(rolling_60day)
```

## 6. Create a pivot table of the FAANG data that compares the stocks. Put the ticker in the rows and show the averages of the OHLC and volume traded data.

```python
import pandas as pd

faang = pd.read_csv('faang.csv')

pivot_table = pd.pivot_table(
    faang,
    values=['open', 'high', 'low', 'close', 'volume'],
    index='ticker',
    aggfunc='mean'
).round(2)

print(pivot_table)
```

## 7. Calculate the Z-scores for each numeric column of Netflix's data (ticker is NFLX) using apply().

```python
import pandas as pd
from scipy import stats

faang = pd.read_csv('faang.csv')

netflix = faang[faang['ticker'] == 'NFLX'].copy()

numeric_cols = ['open', 'high', 'low', 'close', 'volume']

netflix_z_scores = netflix[numeric_cols].apply(stats.zscore)

print("Original Netflix data (first 5 rows):")
print(netflix[numeric_cols].head())
print("\nZ-scores (first 5 rows):")
print(netflix_z_scores.head())
```

## 8. Create a dataframe with the following three columns: ticker, date, and event. 

```python
import pandas as pd

faang = pd.read_csv('faang.csv')
faang['date'] = pd.to_datetime(faang['date'])

event_data = pd.DataFrame({
    'ticker': ['FB', 'FB', 'FB'],
    'date': pd.to_datetime(['2018-07-25', '2018-03-19', '2018-03-20']),
    'event': [
        'Disappointing user growth announced after close.',
        'Cambridge Analytica story',
        'FTC investigation'
    ]
})

event_data = event_data.set_index(['date', 'ticker'])

faang_indexed = faang.set_index(['date', 'ticker'])

merged_data = faang_indexed.merge(event_data, how='outer', left_index=True, right_index=True)

print(merged_data)
```

## 9. Use the transform() method on the FAANG data to represent all the values in terms of the first date in the data. To do so, divide all the values for each ticker by the values for the first date in the data for that ticker. This is referred to as an index, and the data for the first date is the base (https://ec.europa.eu/eurostat/statistics-explained/index.php Beginners:Statisticalconcept-Indexandbaseyear). When data is in this format, we can easily see growth over time. Hint: transform() can take a function name.

```python
import pandas as pd

faang = pd.read_csv('faang.csv')
faang['date'] = pd.to_datetime(faang['date'])

faang = faang.sort_values(['ticker', 'date'])

def calculate_index(group):
    first_values = group.iloc[0]
    numeric_cols = ['open', 'high', 'low', 'close', 'volume']
    group[numeric_cols] = group[numeric_cols].div(first_values[numeric_cols])
    return group

faang_indexed = faang.groupby('ticker').apply(calculate_index)

print("FAANG data indexed to first date (multiply by 100 to get percentage):")
print(faang_indexed.head(10))
```
